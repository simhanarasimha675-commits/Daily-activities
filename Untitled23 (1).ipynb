```python
# --- Basic Input and Output ---
# Note: input(input()) is usually an error; use just input()
n = input("enter a value: ")
print(n)

# --- Conditional Statements ---
n = int(input("enter a number for if-elif-else: "))

# Simple if
if n > 0:
    print("given number is positive")

# if-else
if n > 0:
    print("given number is positive")
else:
    print("given number is negative")

# if-elif-else
if n > 0:
    print("given number is positive")
elif n < 0:
    print("given number is negative")
else:
    print("given number is zero")

# --- Loops ---
# Basic range (start, stop, step)
print("Range 0 to 4:")
for i in range(0, 5, 1):
    print(i)

# User-defined range
n = int(input("enter a number for loop limit: "))
for i in range(0, (n + 1)):
    print(i)

# While condition
print("While loop:")
i = 0
while(i <= n):
    print(i)
    i += 1

# --- Control Flow (pass, continue, break) ---
print("Pass example (does nothing at 3):")
for i in range(5):
    if i == 3:
        pass
    print(i)

print("Continue example (skips 3):")
for i in range(5):
    if i == 3:
        continue
    print(i)

print("Break example (stops before 3):")
for i in range(5):
    if i == 3:
        break
    print(i)

# --- List Operations ---
l = []
print("Empty list:", l)

l = [1, 2, 3, 5, 4]
print("Original list:", l)

l.append(6)
print("After append(6):", l)

l.pop()
print("After pop():", l)

l.sort()
print("After sort():", l)

count_five = l.count(5)
print("Count of number 5:", count_five)
```
