```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})

print(data)

data["age_bin"] = pd.cut(data["age"], bins=3)
print(data)

plt.hist(data["age"], bins=3)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal Width Binning Using pd.cut()")
plt.show()

data["age_frequency_bin"] = pd.qcut(data["age"], q=4)
print(data)

plt.hist(data["age"], bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal Frequency Binning Using pd.qcut()")
plt.show()

bin_min = data["age"].groupby(data["age_frequency_bin"], observed=False).min()
bin_max = data["age"].groupby(data["age_frequency_bin"], observed=False).max()
bin_mean = data["age"].groupby(data["age_frequency_bin"], observed=False).mean()
bin_median = data["age"].groupby(data["age_frequency_bin"], observed=False).median()

print(bin_min)
print(bin_max)
print(bin_mean)
print(bin_median)

custom_bins = [17,25,35,50]
labels = ["Young","Adult","Senior"]
custom_binning = pd.cut(data["age"], bins=custom_bins, labels=labels)
print(custom_binning)

custom_bins = [15,20,25,35,50]
labels = ["Child","Young","Adult","Senior"]
custom_binning = pd.cut(data["age"], bins=custom_bins, labels=labels)
print(custom_binning)

plt.hist(data["age"], bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Custom Binning")
plt.show()
```
