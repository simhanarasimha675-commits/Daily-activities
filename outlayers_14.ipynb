```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Study_Hours": [1,2,3,4,5,6,7,8,20]
})

print(data)
print(data.describe())

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR

outliers_iqr = data[(data < LB) | (data > UB)]
print("IQR Outliers")
print(outliers_iqr.dropna())

z_scores = np.abs(stats.zscore(data["Study_Hours"]))
outliers_z = data[z_scores > 2]
print("Z-Score Outliers")
print(outliers_z)

plt.scatter(range(len(data)), data["Study_Hours"])
plt.xlabel("Index")
plt.ylabel("Study Hours")
plt.title("Scatter Plot Showing Outlier")
plt.show()

plt.hist(data["Study_Hours"], bins=10)
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.title("Histogram Showing Outlier Effect")
plt.show()
```
