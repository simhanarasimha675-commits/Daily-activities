```python
n = int(input("Enter a number: "))

if n > 0:
    print("Given number is positive")
elif n < 0:
    print("Given number is negative")
else:
    print("Given number is zero")

print("\nFor Loop Output:")
for i in range(0, n + 1):
    print(i)

print("\nWhile Loop Output:")
i = 0
while i <= n:
    print(i)
    i += 1

print("\nPASS Example:")
for i in range(5):
    if i == 3:
        pass
    print(i)

print("\nCONTINUE Example:")
for i in range(6):
    if i == 3:
        continue
    print(i)

print("\nBREAK Example:")
for i in range(5):
    if i == 3:
        break
    print(i)

print("\nLIST Example:")

l = []
print("Empty list:", l)

l = [1, 2, 3, 5, 4]
print("List with elements:", l)
```
